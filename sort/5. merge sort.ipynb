### 5.1 병합 정렬 (Merge Sort)

- 최선: `O(N)`

- 최악: `O(Nlog(N))`

배열을 작은 하위 배열로 나눈다. 

각 하위 배열을 정렬한다. 

정렬된 하위 배열을 다시 병합한다.

### 5.3 코드 실습

```python
def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
 
        mergeSort(L)
        mergeSort(R)
 
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
            
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
 
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
```

```python
arr = [12, 11, 13, 5, 6, 7]
print(arr)
mergeSort(arr)
print(arr)
```
