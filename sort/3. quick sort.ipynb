### 3.1. 퀵 정렬 (Quick Sort)

정복 분할(Divide and Conquer) 알고리즘을 사용한다. 핵심 프로세스는 파티셔닝이다. 파티셔닝의 목표는 올바른 위치에 피벗을 배치하는 것이다. 

작은 요소는 피벗의 왼쪽에, 큰 요소는 오른쪽에 이동시킨다. 첫번째 파티셔닝이 완료되면 n번째 파티셔닝도 재귀적으로 수행되어 배열을 정렬한다.

```python
### prep
a = []
n = len(a)


### set variables
L = 0
R = n - 1


### partitioning
def partitioning(a, L, R):
    P = a[R]
    i = L
    for j in range(L, R):
        if a[j] <= P:
            a[i], a[j] = a[j], a[i]            
            i += 1
    a[i], a[R] = a[R], a[i]
    return i


### quick sort
def quicksort(a, L, R):
    if L < R:
        i = partitioning(a, L, R)
        quicksort(a, L, i - 1)
        quicksort(a, i + 1, R)
        

quicksort(a, L, R)
```
