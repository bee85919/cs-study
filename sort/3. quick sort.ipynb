### 3.1.   (Quick Sort)

 (Divide and Conquer)  .   .       . 

   ,    .    n     .

```python
### prep
a = []
n = len(a)


### set variables
L = 0
R = n - 1


### partitioning
def partitioning(a, L, R):
    P = a[R]
    i = L
    for j in range(L, R):
        if a[j] <= P:
            a[i], a[j] = a[j], a[i]            
            i += 1
    a[i], a[R] = a[R], a[i]
    return i


### quick sort
def quicksort(a, L, R):
    if L < R:
        i = partitioning(a, L, R)
        quicksort(a, L, i - 1)
        quicksort(a, i + 1, R)
        

quicksort(a, L, R)
```
