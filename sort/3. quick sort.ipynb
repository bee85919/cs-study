### 3.1. 퀵 정렬 (Quick Sort)

- 평균: `O(nlogn)`

- 최선: `O(logn)`

- 최악: `O(n^2)`

정복 분할 `Divide and Conquer` 알고리즘을 사용한다. 

핵심은 `파티셔닝`이다. `피벗(오른쪽 끝 요소)을 배치`하는 것이다. 

`작은 요소는 피벗의 왼쪽`에, `큰 요소는 피벗의 오른쪽`에 이동된다. 

첫번째 파티셔닝이 완료되면 `나머지 파티셔닝도` `재귀적으로 수행`한다.

### 3.3 퀵 정렬 실습

```python
def partitioning(a, L, H, c):
    c, _a, _p = c+1, a[:], a[H]
    
    p = a[H]
    pi = L-1
    
    for i in range(L, H):
        if a[i] <= p:
            pi += 1
            a[pi], a[i] = a[i], a[pi]            
    a[pi+1], a[H] = a[H], a[pi+1]

    print(f'{c}:\npivot = {_p}\npivot index = {pi+1}\n{_a} → {a}\n')
    return pi+1, c


def quick_sort(a, L, H, c):
    if L < H:
        pi, c = partitioning(a, L, H, c)
        quick_sort(a, L, pi-1, c)
        quick_sort(a, pi+1, H, c)
```

```python
a = [10, 7, 8, 9, 1, 5]
L = 0
H = len(a) - 1
c = 0

print(f'a: {a}\n')
quick_sort(a, L, H, c)
```
